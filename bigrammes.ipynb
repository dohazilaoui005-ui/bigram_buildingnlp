Préparer les données

```python


with open("schekspear_dataset.txt", "r", encoding="utf-8") as f:
    text = f.read()

print("Nombre de caractères :", len(text))

```

tokanisation -> on va adopter la tokanisation par mot car on peut traite chaque 2 mot consecutif pour calculer la proba

```python
import nltk
from nltk import word_tokenize,sent_tokenize
i = 0
sentences_sep = []
sentences = sent_tokenize(text)
for sentence in sentences : 
    sen = "START " + sentence + " END"
    sentences_sep.append(sen)

tokenized_sentences = []
for sentence in sentences_sep:
    # Tokeniser la phrase en mots
    tokens = word_tokenize(sentence, language="french")
    tokenized_sentences.append(tokens)

print(tokenized_sentences[:20])
#maintenat les tokens sont pret 
```

puisque les tokens sont pret avec leur seperateur on va passer a l etape suivante --> former les biagram

```python
biagrams = []
for tokens in tokenized_sentences : 
    for i in range(len(tokens) -1 ): 
        biagram= (tokens[i],tokens[i+1])
        biagrams.append(biagram)

print(biagrams[:50])
```

apres avoir creer les biagrammes on va maintenat faire compter la frequence de chaque biagramme dans tout le txt

```python
from collections import defaultdict

bigrams_count = defaultdict(int)
word_count = defaultdict(int)

for mot1,mot2 in biagrams :
    bigrams_count[(mot1,mot2)] +=1 
    word_count[mot1] +=1 


sorted_bigrams = sorted(bigrams_count.items(), key=lambda x: x[1], reverse=True)
print("Top 12 bigrammes :")
for bigram, count in sorted_bigrams[:20]:
    print(f"{bigram}: {count}")
```

```python
bigrams_proba = defaultdict(int)
for (mot1,mot2),count in bigrams_count.items() :
    bigrams_proba[(mot1,mot2)] = count/word_count[mot1]

print("\nProbabilités de quelques bigrammes :")
for bigram, prob in list(bigrams_proba.items())[:30]:
    print(f"P({bigram[1]} | {bigram[0]}) = {prob:.4f}")
```

ajouter le lissage pour gerer les noms qui n existe pas dans cette dataset 

```python
vocab = set(word_count.keys())
for mot1, mot2 in bigrams_count.keys():
    vocab.add(mot2)
vocab_size = len(vocab)
vocab_size

# Appliquer le lissage de Laplace
smoothed_probs = {}
for (mot1, mot2) in bigrams_count:
    smoothed_probs[(mot1, mot2)] = (bigrams_count[(mot1, mot2)] + 1) / (word_count[mot1] + vocab_size)

# Pour les bigrammes inconnus, la probabilité est :
# P(mot2 | mot1) = 1 / (word_counts[mot1] + vocab_size)

```

pour faciliter la prediction de mot qui suit on va adopter la structure imbrique de dictionaire

```python
# Créer un dictionnaire de transitions
transition_probs = defaultdict(dict)

for (mot1, mot2), prob in smoothed_probs.items():
    transition_probs[mot1][mot2] = prob

# Exemple d'utilisation :
print("\nTransitions depuis 'Je' :")
for mot2, prob in transition_probs["So"].items():
 
        print(f"  P({mot2} | I) = {prob:.4f}")
```

maintenat on va generer

```python
import random

def generate_sentence(transition_probs, max_length=20):
    current_word = "START"
    sentence = [current_word]

    for _ in range(max_length):
        # Récupérer les mots possibles et leurs probabilités
        next_words = transition_probs.get(current_word, {})
        if not next_words:
            break  # Pas de transition possible

        # Choisir le prochain mot aléatoirement selon les probabilités
        next_word = random.choices(
            list(next_words.keys()),
            weights=list(next_words.values()),
            k=1
        )[0]

        sentence.append(next_word)
        current_word = next_word

        # Arrêter si on atteint END
        if current_word == "END":
            break

    # Retirer START et END pour afficher la phrase
    sentence = [word for word in sentence if word not in ["START", "END"]]
    return " ".join(sentence)

# Générer 5 phrases
for _ in range(10):
    print(generate_sentence(transition_probs))
```
